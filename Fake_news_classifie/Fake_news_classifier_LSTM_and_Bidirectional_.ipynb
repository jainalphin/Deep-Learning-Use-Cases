```
import pandas as pd
```

```
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Kaggle Faker News Classifier Using LSTM- Deep LEarning| Natural Language Processing/train.csv")
```

```
df.head()
```

```
df= df.dropna()
```

```
X= df.drop("label",axis=1)
y=df['label']
```

```
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import one_hot

from keras.models import Sequential
from keras.layers import Embedding
from keras.layers import Dense
from keras.layers import LSTM
```

```
## voculabry size
voc_size = 5000
```

# Data Cleaning


```
message = X.copy()
message.reset_index(inplace=True)
message
```

```
import re
import nltk
from nltk.corpus import stopwords

from nltk.stem.porter import PorterStemmer
```

```
nltk.download('stopwords')
```

```
### Dataset Preprocessing
corpus =[]
ps_stem =PorterStemmer()
for i in range(0,len(message)):
  review = re.sub('[^a-zA-Z]',' ',message['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps_stem.stem(word) for word in review if word not in stopwords.words("english")]
  review= ' '.join(review)
  corpus.append(review)
```

```
corpus
```

# One Hot Representation


```
onehot_rep = [one_hot(words,voc_size) for words in corpus]
onehot_rep
```

```
### Padding
sent_length=20
padded_doc = pad_sequences(onehot_rep,maxlen=sent_length,padding="pre")
padded_doc
```

```
len(padded_doc[0])
```

```
len(padded_doc)
```

# Readying Model


```
#dim -- 3000 x 40
embedded_feature_size=40
model= Sequential()
#will embded the words
model.add(Embedding(voc_size,embedded_feature_size,input_length=sent_length))
model.add(LSTM(100))
#output layer
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
print(model.summary())
```

```
## X and y
len(padded_doc),y.shape
```

```
import numpy as np
X_final =np.array(padded_doc)
y_final =np.array(y)
```

```
X_final.shape,y_final.shape
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X_final,y_final,test_size=0.33,random_state=42)
```

# Model Training


```
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```
# under without Dropout

y_pred =model.predict_classes(X_test)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```
from tensorflow.keras.layers import Dropout
```

```
#dim -- 3000 x 40
embedded_feature_size=40
model= Sequential()
#will embded the words
model.add(Embedding(voc_size,embedded_feature_size,input_length=sent_length))
model.add(Dropout(0.7))
model.add(LSTM(100))
model.add(Dropout(0.2))
#output layer
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
print(model.summary())
```

```
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```
# under without Dropout

y_pred =model.predict_classes(X_test)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```
model.save("model.h5")
print("Saved model to disk")
```

# Bi-Directional LSTM RNN

```
from keras.layers import Bidirectional
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import one_hot

from keras.models import Sequential
from keras.layers import Embedding
from keras.layers import Dense
from keras.layers import LSTM
```

```
#dim -- 3000 x 40
embedded_feature_size=40
model1= Sequential()
#will embded the words
model1.add(Embedding(voc_size,embedded_feature_size,input_length=sent_length))
model1.add(Dropout(0.7))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.2))
#output layer
model1.add(Dense(1,activation="sigmoid"))
model1.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
print(model1.summary())
```

```
model1.fit(X_train,y_train,epochs=10,batch_size=40,validation_data=(X_test,y_test))
```

```
model1.predict(X_test)
```

```
# under without Dropout

y_pred =model1.predict_classes(X_test)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
